```python
def is_valid(url):
    """
    Checks whether 'url' is a valid URL.
    """
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)
```

```python
def img_scrap (base_url, ap_link):
    
    site_url = base_url + ap_link
    temp_url = site_url
    
    url_arr = []

    print ('\nSite url: ', site_url)

    site_page = rq.get(site_url)
    time.sleep(20)

    # Parse the site content with html.parser
    site_soup = sp(site_page.content, 'html.parser') 

    # Extracting the img tag
    img_tags = site_soup.find_all('img', class_='s-image')    

    for img_tag in img_tags:

        #src_tag = img_tag.find('src').get_text()
        src_tag = img_tag.attrs.get('src')
        print ('\n url of image:', src_tag)
        if not src_tag:
            # if img does not contain src attribute, just skip
            continue
            #exit        
        url_arr.append(src_tag)

          

    print ('\n Array length:', len(url_arr))
    return (url_arr)
```

```python
def download(d_url, folder_name):
    """
    Downloads a file given an URL and puts it in the folder 'pathname'
    """
    
    direct_name = ''
    if (folder_name == 'saree'):
        direct_name = 'saree'
    elif (folder_name == 'trouser'):
        direct_name = 'trouser'
    else:
        direct_name = 'jeans'
        
    # Path 
    pathname = os.path.join(parent_dir, direct_name)
    
    #print ('\n parent_dir:', parent_dir)
    # if path doesn't exist, make that path dir
    if not os.path.isdir(pathname):
        os.makedirs(pathname)
        
    # download the body of response by chunk, not immediately
    response = rq.get(d_url, stream=True)
    time.sleep(20)
    # get the total file size
    file_size = int(response.headers.get("Content-Length", 0))
    
    # get the file name
    filename = os.path.join(pathname, d_url.split("/")[-1])
    # progress bar, changing the unit to bytes instead of iteration (default by tqdm)
    
    progress = tqdm(response.iter_content(1024), f"Downloading {filename}", total=file_size, unit="B", unit_scale=True, unit_divisor=1024)
    with open(filename, "wb") as f:
        for data in progress:
            # write data read to the file
            f.write(data)
            # update the progress bar manually
            progress.update(len(data))
```

```python
def main(base_url, ap_link, f_name):
    
    imgs = img_scrap(base_url, ap_link)
    
    for img in imgs:
        # for each image, download it
        download(img, f_name)
```

```python
main(base_url, '/saree/s?k=saree', 'saree')
```
