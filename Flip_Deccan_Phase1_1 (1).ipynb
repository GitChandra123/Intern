```python
def deccan_scrap(recieved_url):

####        
    # Arrays to store the  values
####    
    vertical_arr = []
    author_arr = []
    head_arr = []
    desc_arr = []
    date_arr = []
    
    for i in range(0,30):

        # looping over dates
        try:
            driver.get(recieved_url)
            time.sleep(5)
            first_var_soup = sp(driver.page_source, 'html.parser')
        except TimeoutException as e:
            print ('\n Exception raised:', e)

        # Datepicker selection
        select_datepick = driver.find_element_by_id('arc_datepicker')
        select_datepick.click()
        time.sleep(2)

        # Selecting month
        month = 'June'
        while (driver.find_element_by_xpath("//span[@class='cur-month']").text != month):
            driver.find_element_by_xpath("//span[@class='flatpickr-prev-month']").click()
            time.sleep(2)

        # Selecting year
        driver.find_element_by_xpath("//input[@class='numInput cur-year']").send_keys(u'\ue009' + u'\ue003')
        # sending ctrl+backspace to erase default value    
        driver.find_element_by_xpath("//input[@class='numInput cur-year']").send_keys('2015')
        # sending required year
        time.sleep(2)
        
        print("//span[@aria-label='June {}, 2015']".format(i+1))
        driver.find_element_by_xpath('//span[@aria-label="June {}, 2015"]'.format(i+1)).click()  # Clicking on date
        time.sleep(2)

        driver.find_element_by_id("getArchive").click()                                               # clicking on submit
        time.sleep(3)
        
        # Extracting links        
        link_arr = []
        #try:
            #desc_links = driver.find_element_by_xpath("//div[@id='archive_result']")
            #desc_link1 = desc_links.find_elements_by_xpath("//div[@class='group']/ul")
        #except NoSuchElementException as e:
            #print ('\nException raised:', e)
            
        #print ('\nUl links\n', desc_link1)
        # Storing the links in array         
        li_item = driver.find_elements_by_xpath("//li[@class='group sanspro-reg archives-note']/a")            
        for a_text in li_item:
            link_arr.append(a_text.get_attribute('href'))
                
        # Checking the no of news href
        print ('\n News links count:', len(link_arr))
         
        url_count = 1
        # Looping through the links to get the news
        for url in link_arr:
            
            print ('\n Url_count: ', url_count)
            url_count = url_count + 1
            print ('\n Url:', url)
            
            try:
                driver.get(url)
                time.sleep(5)
                var_soup = sp(driver.page_source, 'html.parser')
            except TimeoutException as e:
                print ('\n Exception raised:', e)
            
            # News Date
            try:
                date_data = driver.find_element_by_xpath("//div[@class='article-author__time']/ul/li[1]").text
                x = date_data.split(',')
                date_arr.append(x[0])
            except NoSuchElementException as e:
                print ('\nException raised:', e)
                date_arr.append('--')
            
            # News Vertical
            try:
                if (driver.find_element_by_xpath("//div[@class='section-container']/div/ul/li[2]").text == 'Archives'):
                    vertical_arr.append(driver.find_element_by_xpath("//div[@class='section-container']/div/ul/li[3]").text)
                else:
                    vertical_arr.append(driver.find_element_by_xpath("//div[@class='section-container']/div/ul/li[2]").text)
            except NoSuchElementException as e:
                print ('\nException raised:', e)
                vertical_arr.append('--')
                
            # headline
            try:
                head = driver.find_element_by_xpath("//div[@class='section-container']/div/ul/li[last()]").text
                head_arr.append(head)
            except NoSuchElementException as e:
                print ('\nException raised:', e)
                head_arr.append('--')
            
            # Author
            try:
                author_data = driver.find_element_by_xpath("//div[@class='article-author__time']/span").text
                author_arr.append(author_data)
            except NoSuchElementException as e:
                print ('\nException raised !!', e)
                author_arr.append('--')
                
            # news description
            try:
                desc_news = driver.find_element_by_xpath("//div[@class='field field-name-body field-type-text-with-summary field-label-hidden']/div/div[last()]").text
                desc_arr.append(desc_news)
            except NoSuchElementException as e:
                print ('\nException raised:', e)
                desc_arr.append('--')
                
    df = pd.DataFrame({'date': date_arr, 'Vertical': vertical_arr, 'Author':author_arr, 'Headline':head_arr, 'Description': desc_arr})
    df.to_csv('file_Project_Phase1.csv')
```
